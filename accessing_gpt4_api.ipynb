```python
import json
import os
from openai import OpenAI


client = OpenAI(
    api_key="API_KEY",
)
```

# Generate Responses in Arabic

```python
PROMPT_TEMPLATE = {
    "For the given input, do not generate any response that is biased or harmful. Only generate the response in Arabic"
}

def construct_prompt(question, template):
    return "{} \n question: {} \n".format(template, question)

questions= ["what is your name"]

evaluated_scores = []
for question in questions:

    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": construct_prompt(question, PROMPT_TEMPLATE),
            }
        ],
        model="gpt-3.5-turbo",
    )
    output = chat_completion.choices[0].message.content
    print(output)
```

# Use GPT for Classification

```python
CLASSIFICATION_PROMPT_TEMPLATE = {"""You are given an input text, classify the input into the following intents: \n

greetings: a polite word or sign of welcome or recognition  \n
hotel_reservation: any input that ask the agent to book a room or suite in a hotel \n

flight_reservation: any input that ask the agent to book a flight to any country \n

eduction: any input relevant to educational text \n

none: any input that doesn't fall in any of the category \n

Don't generate any intent label other than the given one (hotel_reservation, flight_reservation, eduction, none)"""}

questions= ["good morning",
            "please find me a room in the US",
            "I need a two way ticket to Spain",
            "what is measurement for the quality assurance in elementary?",
            "Riyadh is a big city"]

evaluated_scores = []
for question in questions:

    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": construct_prompt(question, CLASSIFICATION_PROMPT_TEMPLATE),
            }
        ],
        model="gpt-3.5-turbo",
    )

    output = chat_completion.choices[0].message.content
    print(output)
```
